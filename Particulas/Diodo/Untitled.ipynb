```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('GP1_2911_scan_electrones_current.txt') as fr:
    line = fr.readline()
    voltage, inten = [], []
    while line != '':
        vol, inte = line.split('\t')
        voltage.append(float(vol))
        inten.append(float(inte))
        line = fr.readline()
plt.plot(inten, voltage)
plt.show()
```

```python
with open('GP2_2911_scan_grueso_electrones.dat') as fr:
    t0 = float(fr.readline().split(':')[1][:-3])
    dt = float(fr.readline().split(':')[1][:-3])
    samples = int(fr.readline().split(':')[1][:-3])
    volts = fr.readline().split('\t')
    
    for i in range(len(volts)):
        volts[i] = 1000/float(volts[i])
    
    perfiles = [[] for i in range(len(volts))]
    
    line = fr.readline()
    while line != '':
        valores = line.split('\t')
        for i in range(len(perfiles)):
            perfiles[i].append(float(valores[i]))
        line = fr.readline()
    perfiles.append([t0+dt*i for i in range(samples)])
    
i = 1
for values in perfiles[i:-1]:
    plt.plot(perfiles[-1], values, label='%.2f mV' %(volts[i]))
    i+=1
plt.legend()
plt.show()
```
