```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('GP1_2911_scan_electrones_current.txt') as fr:
    line = fr.readline()
    voltage, inten = [], []
    while line != '':
        vol, inte = line.split('\t')
        voltage.append(float(vol))
        inten.append(float(inte))
        line = fr.readline()
fig = plt.figure()
plt.plot(voltage, inten, 'o')
plt.xscale('log')
plt.xlabel('$V_{bias}$/V', fontsize=15)
plt.ylabel('Corriente Inversa/$A$', fontsize=15)
plt.show()
fig.savefig("todos.png", dpi=300)
```

```python
indexVolt = voltage.index(100)

volts = voltage[:indexVolt]
iInt = inten[:indexVolt]

print np.mean(iInt), np.std(iInt)
```

```python
with open('GP2_2911_scan_grueso_electrones.dat') as fr:
    t0 = float(fr.readline().split(':')[1][:-3])
    dt = float(fr.readline().split(':')[1][:-3])
    samples = int(fr.readline().split(':')[1][:-3])
    volts = fr.readline().split('\t')
    
    for i in range(len(volts)):
        volts[i] = float(volts[i])
    
    perfiles = [[] for i in range(len(volts))]
    
    line = fr.readline()
    while line != '':
        valores = line.split('\t')
        for i in range(len(perfiles)):
            perfiles[i].append(float(valores[i]))
        line = fr.readline()
    perfiles.append([t0+dt*i for i in range(samples)])
    
i = 1
for values in perfiles[i:-1]:
    plt.plot(perfiles[-1], values, label='%.2f V' %(volts[i]))
    i+=1
plt.legend()
plt.show()
```

```python
with open('GP2_2911_scan_fino_electrones.dat') as fr:
    t0 = float(fr.readline().split(':')[1][:-3])
    dt = float(fr.readline().split(':')[1][:-3])
    samples = int(fr.readline().split(':')[1][:-3])
    volts = fr.readline().split('\t')
    
    for i in range(len(volts)):
        volts[i] = float(volts[i])
    
    perfiles = [[] for i in range(len(volts))]
    
    line = fr.readline()
    while line != '':
        valores = line.split('\t')
        for i in range(len(perfiles)):
            perfiles[i].append(float(valores[i]))
        line = fr.readline()
    perfiles.append([t0+dt*i for i in range(samples)])

fig = plt.figure()
plt.ylabel('$V_{inducida}$/V', fontsize=15)
plt.xlabel('t/s', fontsize=15)
i = 0
for values in perfiles[i:-1]:
    plt.plot(perfiles[-1], values, label='%.2f V' %(volts[i]))
    i+=1
plt.legend()
plt.show()
fig.savefig("fina.png")
```

```python
fig = plt.figure()
plt.ylabel('$V_{inducida}$/V', fontsize=15)
plt.xlabel('t/s', fontsize=15)
plt.plot(perfiles[-1], perfiles[2], "r")
plt.show()
fig.savefig("Ejemplo.png")
```

```python
fig = plt.figure()
plt.xlabel('$V_{bias}$/V', fontsize=15)
plt.ylabel('Q/C', fontsize=15)

plt.legend()

I = [0 for i in range(len(volts))]
trapz = [0 for i in range(len(volts))]
i = 0
for j in range(len(perfiles[i:-1])):
    perfiles[j] = np.asarray(perfiles[j])
    I[j] = perfiles[j]/50
    trapz[j] = np.trapz(I[j], perfiles[-1])
    
plt.plot(volts, trapz)    
plt.show()
fig.savefig("Q.png")

e = -1.60217656535*10**(-19)
print (np.mean(trapz))/(2*e*1000), np.std(trapz)/(e*1000)
```
